```python
def process_text(text):
    text = str(text)
    text = re.sub(r'[^\w\s]','',text)
    text = text.lower()
    text = text.replace('\n',' ').replace('\t' , ' ')
    text = text.strip()
    return text.split()
```

```python
def minmax_normal(score,negative = False):
    if negative:
        return (score - score.min()) / (score.max() - score.min())
    else:
        return 2 * (score - score.min()) / (score.max() - score.min()) -1
```

```python
def print_result(bm25_score,model_score,query_bm25_doc_id,query_id , alpha = alpha):
    print(query_id + ',', file=fp, end='')

    scores = bm25_score + alpha * model_score

    doc_dict = dict(zip(query_bm25_doc_id, scores))
    sorted_docs = sorted(doc_dict.items(), key=lambda x: x[1], reverse=True)
    print(sorted_docs[:20])

    for _doc in sorted_docs:
        doc_id, value = _doc
        print(doc_id, file=fp, end=' ')
    print('',file=fp)
```

```python
# relate is label
def get_map_score(q_result , q_relate):
    _map = 0.0
    for i in range(len(q_result)):
        ap = 0.0
        results = q_result[i]
        relates = q_relate[i]
        idx = 0
        count = 0
        for result in results:
            if result in relates:
                count += 1
                ap += count / (idx + 1)
            idx += 1


        ap /= len(relates)
        _map += ap

    _map = round(_map/len(q_result),4)
    return _map
```

```python
def get_topk_doc(bm25_score, model_score, query_bm25_doc_id, query_id ,alpha = alpha):
    scores = bm25_score + alpha * model_score
    doc_dict = dict(zip(query_bm25_doc_id, scores))
    sorted_docs = sorted(doc_dict.items(), key=lambda x: x[1], reverse=True)
    return sorted_docs
```

```python
from torch.utils.data import Dataset
from torch.utils.data import DataLoader

class TestDataset(Dataset):
    def __init__(self, input_ids, token_type_ids , attention_mask):
        self.input_ids = input_ids
        self.token_type_ids = token_type_ids
        self.attention_mask = attention_mask        
    def __getitem__(self,idx):
        inputid = self.input_ids[idx]
        tokentype = self.token_type_ids[idx]
        attentionmask = self.attention_mask[idx]
        return inputid , tokentype , attentionmask
    def __len__(self):
        return len(self.input_ids)
    
```

```python
def softmax(x):
    """Compute the softmax in a numerically stable way."""
    x = x - np.max(x)
    exp_x = np.exp(x)
    softmax_x = exp_x / np.sum(exp_x)
    return softmax_x
def normal(vec):
    if np.sum(vec) > 0:
        return vec / np.sum(vec)
    else:
        return vec
```
